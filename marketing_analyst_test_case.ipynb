Task 3: Reviewing Ecommerce Product Data
Objective: Test attention to detail by identifying product performance trends and making strategic decisions.
Context: You are a Junior Marketing Analyst at a medium-sized e-commerce store specializing in home electronics. Due to increased costs from suppliers, the company implemented a price increase across all products at the beginning of Month 2. The Marketing Director is concerned about how these price adjustments impacted key performance indicators (KPIs) such as views, conversions, and revenue.
Task:
Your task is to analyze the provided dataset, comparing product performance between Month 1 (before price increase) and Month 2 (after price increase). Use the table below to identify patterns and prepare actionable insights. Consider how the price increase might have influenced consumer behavior and product performance.
Identify products most and least impacted by the price increase.
Highlight noticeable changes in conversion rates (View-to-Purchase and Cart-to-Purchase).
Recommend actions based on your findings.


```python
import pandas as pd
file_path = '/content/drive/MyDrive/Practical task google analytics/Лист Microsoft Excel.xlsx'
ecom_df = pd.read_excel(file_path)
ecom_df.head(5)
```

Из предоставленных данных видно, что колличество просмотров проудктов магазина припало в сравнении первого месяца ко второму за исключением двух продуктов (Smart Home Hub, Wireless Mouse) что может быть обусловленно иными изменения рынка.


```python
# ecom_df.dtypes
```

```python
ecom_df['View-to-Purchase Rate (%)'] = ecom_df['View-to-Purchase Rate (%)'].str.replace('%', '').str.strip()
ecom_df['View-to-Purchase Rate (%)'] = ecom_df['View-to-Purchase Rate (%)'].astype(float)
ecom_df['Cart-to-Purchase Rate (%)'] = ecom_df['Cart-to-Purchase Rate (%)'].str.replace('%', '').str.strip()
ecom_df['Cart-to-Purchase Rate (%)'] = ecom_df['Cart-to-Purchase Rate (%)'].astype(float)
```

```python
print(ecom_df.columns)
```

```python
first_month = ecom_df[ecom_df['Month'] == 1]
sec_month = ecom_df[ecom_df['Month'] == 2]
```

```python
view2purch_df = pd.merge(first_month, sec_month, on='Product Name')
view2purch_df.head(2)
```

```python
view2purch_df['view_delta'] = view2purch_df['Item Views_x'] - view2purch_df['Item Views_y']
view_delta_df = view2purch_df.sort_values('view_delta', ascending=False)
view_delta_df.tail(2)
```

```python
difference_views = view2purch_df['Item Views_y'].sum() - view2purch_df['Item Views_x'].sum()
print(difference_views)
```

```python
difference_rev = view2purch_df['Revenue ($)_y'].sum() - view2purch_df['Revenue ($)_x'].sum()
print(difference_rev * 1000)
```

**Топ 2 товара с отрицательным ростом конверсии просмотр/покупка это:**

Gaming Headset ( 8.8% )
Smartwatch ( 4.6% )

```python
view2purch_df['delta_view2purchase'] = view2purch_df['View-to-Purchase Rate (%)_x'] - view2purch_df['View-to-Purchase Rate (%)_y']
top_delta_df = view2purch_df.sort_values('delta_view2purchase', ascending=False)
top_delta_df.head(2)
```

**Топ 2 товара с положительным ростом конверсии просмотр/покупка это:**

Bluetooth Speaker ( 0.1% )
Portable Charger ( 0.2% )

```python
top_delta_df.tail(2)
```

```python
view2purch_df['delta_card2purchase'] = view2purch_df['Cart-to-Purchase Rate (%)_x'] - view2purch_df['Cart-to-Purchase Rate (%)_y']
top_card_delta_df = view2purch_df.sort_values('delta_card2purchase', ascending=False)
top_card_delta_df.head(2)
```

```python
top_card_delta_df.tail(2)
```

Мы выяснили что наши продукты имеют разный уровень гибкости спроса и по-разному откликаются спросом на колебания цены. Нам необходимо оптимизировать цены и зарабатывать больше на товарах с низкой гибкостью спроса, чтобы компенсировать общее падение выручки. На фоне общего подорожания мы потеряли в выручке и нам необходимо проверить гипотезы как в сторону снижения цен для формирования больших обьемов выручки ща счет нашей марженальности, так и в сторону повышения цен, если емкость рынка не позволит нам увеличить прибыль увеличив выручку в достаточной мере и нам придется увеличивать марженальность.
